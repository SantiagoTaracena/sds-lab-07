### Primera parte

En la primera parte del proyecto, se debe desarrollar un modelo de Deep Learning para clasificar imágenes de malware. Este proceso comienza con la descarga y el preprocesamiento de un conjunto de datos, compuesto por imágenes PNG que representan 25 diferentes familias de malware. Se debe asegurar una distribución adecuada de imágenes por familia, posiblemente descartando aquellas con pocos ejemplos para evitar sesgos en el entrenamiento. Luego, se procederá a diseñar y configurar la arquitectura de la red neuronal, seleccionando las capas, funciones de activación y optimizadores adecuados. El modelo se entrenará utilizando un 70% de los datos para el entrenamiento y el 30% restante para pruebas, cuidando de evitar el sobreajuste y seleccionando el número de épocas óptimo. Finalmente, el modelo será evaluado usando métricas apropiadas para medir su precisión y generalización, y se guardará para su uso en la segunda parte del proyecto.

```python
# Librerías necesarias para la primera parte
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Path a utilizar para cargar las imágenes
path = "./data/malimg_dataset/malimg_paper_dataset_imgs/"
path
```

```python
# Familias obtenidas del directorio especificado
families = ImageDataGenerator().flow_from_directory(directory=path, target_size=(64, 64), batch_size=10_000)
```

```python
# Índices correspondientes a las familias encontradas
families.class_indices
```

```python
# Imágenes y categorías de las familias.
images, labels = next(families)
```

```python
# Tupla con información sobre el dataset de imágenes.
images.shape
```

```python
# Información sobre imágenes y categorías.
labels.shape
```

```python
# plots images with labels within jupyter notebook
def plots(images, figsize=(20, 30), rows=10, interp=False, titles=None):

    if (type(images[0]) is np.ndarray):
        images = np.array(images).astype(np.uint8)
        if (images.shape[-1] != 3):
            images = images.transpose((0,2,3,1))

    figure = plt.figure(figsize=figsize)
    cols = 10 # len(ims)//rows if len(ims) % 2 == 0 else len(ims)//rows + 1

    for i in range(0, 50):
        sp = figure.add_subplot(rows, cols, i+1)
        sp.axis("Off")
        if (titles is not None):
            sp.set_title(list(families.class_indices.keys())[np.argmax(titles[i])], fontsize=16)
        plt.imshow(images[i], interpolation=None if (interp) else "none")
```

```python
# Graficar imágenes del dataset
plots(images, titles=labels)
```

```python
# Librerías necesarias para el desarrollo del modelo
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
# Familias a tomar en cuenta
families = {}

# Iteración y búsqueda de clases
for folder in os.listdir(path):
    if os.path.isdir(os.path.join(path, folder)):
        count = len(os.listdir(os.path.join(path, folder)))
        families[folder] = count

# Retorno de las familias encontradas
families
```

```python
# Filtración de familias sin datos de peso
filtered_families = {k: v for k, v in families.items() if (v >= 100)}
filtered_families
```

```python
# Familias restantes luego de la filtración
len(filtered_families)
```

```python
# Cuentas de familias filtradas
plt.bar(filtered_families.keys(), filtered_families.values())
plt.xticks(rotation=90)
plt.xlabel("Familia de Malware")
plt.ylabel("Número de Imágenes")
plt.show()
```

```python
# Imágenes normalizadas
normalized_images = images / 255.0
```

```python
# LabelBinarizer y train_test_split para transformar la data
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
```

```python
# Instancia de LabelBinarizer
label_binarizer = LabelBinarizer()
label_binarizer
```

```python
# Labels codificados para cada imagen
encoded_labels = label_binarizer.fit_transform(labels)
encoded_labels
```

```python
# Separación del dataset en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(normalized_images, encoded_labels, test_size=0.3, random_state=42)
```

```python
# Instancia del modelo a crear para la clasificación
model = Sequential([
    Conv2D(32, (3, 3), activation="relu", input_shape=(64, 64, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation="relu"),
    Dropout(0.5),
    Dense(len(families), activation="softmax")
])
model
```

```python
# Compilación del modelo
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
# Entrenamiento del modelo
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_data=(X_test, y_test),
)
```

```python
# Graficar la evolución del entrenamiento
plt.plot(history.history["accuracy"], label="Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.legend()
plt.show()
```

```python
# Pérdida y precisión del modelo entrenado
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

```python
# Almacenamiento del modelo entrenado
model.save("./models/malware_classifier_model.h5")
```

### Segunda parte

En la segunda parte del proyecto, se evaluará la robustez del modelo de clasificación de malware desarrollado anteriormente mediante la implementación de dos ataques adversarios utilizando el Adversarial Robustness Toolbox (ART). El primer ataque será un ataque de evasión, donde se generarán ejemplos adversarios para intentar engañar al modelo y provocar clasificaciones incorrectas. El segundo será un ataque de envenenamiento, donde se introducirán datos maliciosos durante el entrenamiento para alterar la efectividad del modelo. Cada ataque se documentará detalladamente, explicando la metodología y la implementación práctica, y se evaluarán los resultados para entender cómo el modelo maneja estas intervenciones adversas. Esta fase es crucial para identificar vulnerabilidades potenciales en el modelo y para mejorar las medidas de seguridad en futuros desarrollos.

El primer ataque a realizar es un ataque FGSM. El ataque de Fast Gradient Sign Method (FGSM) es un método de ejemplo adversario diseñado para engañar a los modelos de aprendizaje profundo, especialmente en tareas de clasificación. Desarrollado por Ian Goodfellow y sus colaboradores en 2014, FGSM es un ataque de caja blanca, lo que significa que asume acceso al modelo objetivo, incluyendo su arquitectura y gradientes.

La premisa básica del FGSM es simple: utiliza el gradiente del modelo respecto a la entrada para crear una nueva imagen que sea casi indistinguible para el ojo humano pero que resulte en una clasificación incorrecta por parte del modelo. El objetivo del ataque es maximizar el error de predicción alterando ligeramente la imagen de entrada en la dirección del gradiente de la función de pérdida respecto a la entrada.

Para implementar FGSM, se calcula primero el gradiente de la función de pérdida del modelo respecto a la entrada (por ejemplo, una imagen). Este gradiente indica en qué dirección debería cambiar cada pixel de la imagen para aumentar la pérdida, lo que probablemente haría que el modelo cometa un error en la clasificación. En lugar de ajustar los píxeles de la imagen de manera proporcional a este gradiente, FGSM toma el signo del gradiente y lo multiplica por un factor épsilon (𝜖), un pequeño valor escalar. Este proceso genera una perturbación que se añade a la imagen original para crear la imagen adversaria.

El factor ϵ juega un papel crucial en FGSM. Controla la magnitud de la perturbación y, por ende, la efectividad del ataque. Si ϵ es muy pequeño, la perturbación puede no ser suficiente para engañar al modelo. Por otro lado, un 𝜖 demasiado grande puede hacer que las perturbaciones sean fácilmente perceptibles, lo que podría alertar sobre un posible ataque o simplemente hacer que la imagen resultante sea irreconocible.

FGSM es particularmente famoso por su eficiencia y simplicidad, ya que requiere una única pasada de cálculo de gradientes, lo que lo hace mucho más rápido que otros métodos de generación de ejemplos adversarios que podrían requerir iteraciones múltiples. Esto hace a FGSM un ataque práctico y potente para evaluar la robustez de los modelos de aprendizaje profundo en situaciones donde el adversario tiene un conocimiento completo del modelo y necesita generar ataques rápidamente. Sin embargo, la simplicidad de FGSM también puede ser una desventaja, ya que los ataques más sofisticados pueden requerir enfoques más elaborados que consideren múltiples pasos o ajustes más finos.

```python
# Función para cargar el modelo creado
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
# Desactivar la ejecución ansiosa
tf.compat.v1.disable_eager_execution()
```

```python
# Cargar el modelo preentrenado
model = load_model("./models/malware_classifier_model.h5")
model
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(normalized_images, encoded_labels, test_size=0.3, random_state=42)
```

```python
# Librerías necesarias para el primer ataque
from art.estimators.classification import KerasClassifier
from art.attacks.evasion import FastGradientMethod
```

```python
# Conversión del modelo a uno de clasificación de ART
art_model = KerasClassifier(model=model, clip_values=(0, 1))
art_model
```

```python
# Instancia del ataque
fgsm_attack = FastGradientMethod(estimator=art_model, eps=0.1)
fgsm_attack
```

```python
# Generación de ejemplos adversarios con FGSM
X_test_adv_fgsm = fgsm_attack.generate(x=X_test)
X_test_adv_fgsm
```

```python
# Predicciones del modelo
predictions = art_model.predict(X_test_adv_fgsm)
predictions
```

```python
# Precisión del modelo obtenido
accuracy = np.sum(np.argmax(predictions, axis=1) == np.argmax(y_test, axis=1)) / len(y_test)
accuracy
```

Un ataque FGSM (Fast Gradient Sign Method) con un accuracy de 0.17701641684511063 nos da a entender que hemos obtenido un modelo de clasificación indica que el modelo solo pudo clasificar correctamente aproximadamente el 17.7% de las entradas modificadas mediante este método de ataque. Esto sugiere que el ataque fue altamente efectivo en engañar al modelo, provocando un alto grado de error en las predicciones.

El siguiente ataque a realizar es un ataque DeepFool. Un ataque DeepFool es un método de generación de ejemplos adversarios diseñado para engañar a modelos de aprendizaje automático, especialmente en tareas de clasificación. Desarrollado por Moosavi-Dezfooli y colegas en 2016, DeepFool es reconocido por su enfoque iterativo y su eficacia para encontrar la mínima perturbación necesaria para inducir a un error de clasificación. A diferencia del ataque FGSM, que realiza una única perturbación grande y fija, DeepFool ajusta meticulosamente la perturbación en varias iteraciones para hacerla tan pequeña e indetectable como sea posible.

El objetivo principal de DeepFool es calcular la perturbación mínima que cambia la clasificación de una entrada dada en un clasificador lineal, y luego extender este enfoque a clasificadores más complejos como redes neuronales profundas. Para lograr esto, DeepFool considera el problema de clasificación como un problema de optimización geométrica, donde se busca la menor distancia (perturbación) que mueva una muestra desde su posición original hacia el límite de decisión más cercano en el espacio de entrada.

El ataque DeepFool comienza con una imagen original que el modelo clasifica correctamente. Luego, en cada iteración, el ataque estima linealmente el límite de decisión del clasificador basado en el gradiente de la función de pérdida respecto a la imagen. Utilizando esta aproximación lineal, se calcula la perturbación más pequeña que movería la imagen a través de este límite de decisión. La imagen se modifica entonces ligeramente en esta dirección y el proceso se repite hasta que la imagen es mal clasificada por el modelo o hasta que se alcanza un número máximo de iteraciones.

Una de las ventajas clave de DeepFool es su eficiencia en encontrar la perturbación más pequeña necesaria para engañar al modelo, lo cual es particularmente útil en escenarios donde se desea que los ejemplos adversarios sean imperceptibles al ojo humano. Además, la metodología iterativa permite una exploración más precisa y efectiva del espacio de entrada en comparación con métodos que utilizan una única perturbación grande, como FGSM.

DeepFool ha demostrado ser efectivo en una variedad de tareas de clasificación y es ampliamente utilizado para evaluar la robustez de modelos en la visión por computadora y otros campos del aprendizaje automático. Sin embargo, debido a que su enfoque es más computacionalmente intenso que métodos como FGSM, puede no ser adecuado para todas las aplicaciones, especialmente en entornos donde la velocidad de ejecución es crítica.

```python
# Importar la clase que permite un ataque DeepFool
from art.attacks.evasion import DeepFool
```

```python
# Instancia de un ataque DeepFool
deepfool_attack = DeepFool(art_model)
deepfool_attack
```

```python
# Generación de ejemplos adversarios con DeepFool
X_test_adv_deepfool = deepfool_attack.generate(x=X_test)
X_test_adv_deepfool
```

```python
# Predicciones del modelo DeepFool
predictions = art_model.predict(X_test_adv_deepfool)
predictions
```

```python
# Precisión del ataque
accuracy = np.sum(np.argmax(predictions, axis=1) == np.argmax(y_test, axis=1)) / len(y_test)
accuracy
```

Un ataque DeepFool que resulta en una precisión del modelo de tan solo 0.04318344039971449 (aproximadamente 4.3%) indica un nivel extremadamente alto de eficacia en la generación de imágenes adversarias que engañan al modelo. Esto significa que casi todas las imágenes manipuladas mediante DeepFool fueron clasificadas incorrectamente por el modelo, subrayando así una vulnerabilidad significativa en la capacidad del modelo para manejar perturbaciones mínimas.
